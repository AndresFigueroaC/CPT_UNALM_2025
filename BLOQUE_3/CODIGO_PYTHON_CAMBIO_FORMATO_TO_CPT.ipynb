```python
import xarray as xr
import pandas as pd

#SUBIENDO DATOS
ruta_archivo = r"D:\Documents_2\LABORAL\2025\CPT_2025\CURSO_UNALM_CPT_2025\BLOQUE_3\DATA\pisco_mpr_v2.1__updated2019.nc"
ds = xr.open_dataset(ruta_archivo)

# ASIGNANDO COORDENADA
n_times = ds.sizes['z']  # tamaÃ±o real de la dimensiÃ³n 'z' (compatible con futuras versiones)
start_date = pd.Timestamp('1981-01-01')
dates = pd.date_range(start=start_date, periods=n_times, freq='M')
ds = ds.assign_coords(time=("z", dates)).swap_dims({"z": "time"})
if 'crs' in ds.variables:
    ds = ds.drop_vars('crs')
ds = ds.resample(time='MS').sum()
ds
```

```python
import xarray as xr
import numpy as np
import os

# CAMBIOS QUE SE USARAN 
dataset ='pisco'
varname = "pc"                          # Nombre de la variable en el dataset
op = "sum" # "mean" o "sum"
months = [1, 2, 3]                    # Lista de meses (1=Ene, 12=Dic)

# ZONA SUR
#lat_range = (-14, -18.5)  # (lat_norte, lat_sur)
#lon_range = (-75, -69)    # (lon_oeste, lon_este)

#ZONA NORTE
lat_range = (-3, -8)     # (lat_norte, lat_sur)
lon_range = (-81, -75)   # (lon_oeste, lon_este)


units = "mm/month"                      # Unidades de salida
missing = -999                          # Valor para missing data
standard_name = "lwe_precipitation_rate"# CF standard_name
output_dir = R"D:\Documents_2\LABORAL\2025\CPT_2025\CURSO_UNALM_CPT_2025\BLOQUE_3\CONVERSION"        # Carpeta donde guardar archivo
ds = ds


# CAMBIANDO A FORMATO CPT
precip = ds[varname]

precip = precip.sel(lat=slice(*lat_range), lon=slice(*lon_range))
precip_sel = precip.sel(time=precip['time'].dt.month.isin(months))
if op == "mean":
    precip_sel = precip_sel.groupby('time.year').mean('time')
elif op == "sum":
    precip_sel = precip_sel.groupby('time.year').sum('time')
else:
    raise ValueError("op debe ser 'mean' o 'sum'")
if precip_sel.lat.values[0] < precip_sel.lat.values[-1]:
    precip_sel = precip_sel.sortby('lat', ascending=False)
lats = precip_sel.lat.values
lons = precip_sel.lon.values
years = precip_sel.year.values
header = "xmlns:cpt=http://iri.columbia.edu/CPT/v10/\n"
header += "xmlns:cf=http://cf-pcmdi.llnl.gov/documents/cf-conventions/1.4/\n"
header += "cpt:nfields=1\n"

if len(months) > 1:
    periodos = [f"{y}-{months[0]:02d}/{months[-1]:02d}" for y in years]
else:
    periodos = [f"{y}-{months[0]:02d}" for y in years]

header += "cpt:T\t" + "\t".join(periodos) + "\n"
bloques = []
for i, periodo in enumerate(periodos):
    meta = (f"cpt:field={varname}, cpt:T={periodo}, "
            f"cpt:nrow={len(lats)}, cpt:ncol={len(lons)}, "
            f"cpt:row=Y, cpt:col=X, cpt:units={units}, "
            f"cf:standard_name={standard_name}, cpt:missing={missing}\n\t" +
            "\t".join(f"{lon:.2f}" for lon in lons) + "\n")
    data = precip_sel.isel(year=i).values
    for lat_idx, lat in enumerate(lats):
        row = data[lat_idx, :]
        row_str = "\t".join(f"{v:.4f}" if np.isfinite(v) else str(missing) for v in row)
        meta += f"{lat:.4f}\t{row_str}\n"
    bloques.append(meta)

# === 9. Guardar ===
os.makedirs(output_dir, exist_ok=True)
fname = os.path.join(output_dir, f"Y_{dataset}_{varname}_{'_'.join(str(m) for m in months)}_{op}.tsv")

with open(fname, "w") as f:
    f.write(header)
    for bloque in bloques:
        f.write(bloque)

print(f"Archivo guardado en: {fname}")
```

```python
# CONVERSION "m" a "mm"
ds['sro'] = ds['sro']*1000
ds['sro'].attrs['units'] = 'mm (precipitaciÃ³n equivalente)'
ds
```

```python
import xarray as xr
import numpy as np
import os

# APLICANDO CAMBIOS
dataset ='ERA5'
varname = "sro"                          # Nombre de la variable en el dataset
op = "sum" # "mean" o "sum"
months = [9,10, 11]                    # Lista de meses (1=Ene, 12=Dic)

# ZONA SUR
#lat_range = (-14, -18.5)  # (lat_norte, lat_sur)
#lon_range = (-75, -69)    # (lon_oeste, lon_este)

#ZONA NORTE
lat_range = (-3, -8)     # (lat_norte, lat_sur)
lon_range = (-81, -75)   # (lon_oeste, lon_este)

units = "mm"                      # Unidades de salida
missing = -999                          # Valor para missing data
standard_name = "surface_runoff"# CF standard_name
output_dir = R"D:\Documents_2\LABORAL\2025\CPT_2025\CURSO_UNALM_CPT_2025\BLOQUE_3\CONVERSION"        # Carpeta donde guardar archivo
ds = ds

# CAMBIANDO A FORMATO CPT
precip = ds[varname]
precip = precip.sel(lat=slice(*lat_range), lon=slice(*lon_range))
precip_sel = precip.sel(time=precip['time'].dt.month.isin(months))
if op == "mean":
    precip_sel = precip_sel.groupby('time.year').mean('time')
elif op == "sum":
    precip_sel = precip_sel.groupby('time.year').sum('time')
else:
    raise ValueError("op debe ser 'mean' o 'sum'")
if precip_sel.lat.values[0] < precip_sel.lat.values[-1]:
    precip_sel = precip_sel.sortby('lat', ascending=False)
lats = precip_sel.lat.values
lons = precip_sel.lon.values
years = precip_sel.year.values
header = "xmlns:cpt=http://iri.columbia.edu/CPT/v10/\n"
header += "xmlns:cf=http://cf-pcmdi.llnl.gov/documents/cf-conventions/1.4/\n"
header += "cpt:nfields=1\n"

if len(months) > 1:
    periodos = [f"{y}-{months[0]:02d}/{months[-1]:02d}" for y in years]
else:
    periodos = [f"{y}-{months[0]:02d}" for y in years]

header += "cpt:T\t" + "\t".join(periodos) + "\n"

# === 8. Crear bloques ===
bloques = []
for i, periodo in enumerate(periodos):
    meta = (f"cpt:field={varname}, cpt:T={periodo}, "
            f"cpt:nrow={len(lats)}, cpt:ncol={len(lons)}, "
            f"cpt:row=Y, cpt:col=X, cpt:units={units}, "
            f"cf:standard_name={standard_name}, cpt:missing={missing}\n\t" +
            "\t".join(f"{lon:.2f}" for lon in lons) + "\n")
    data = precip_sel.isel(year=i).values
    for lat_idx, lat in enumerate(lats):
        row = data[lat_idx, :]
        row_str = "\t".join(f"{v:.4f}" if np.isfinite(v) else str(missing) for v in row)
        meta += f"{lat:.4f}\t{row_str}\n"
    bloques.append(meta)

# === 9. Guardar ===
os.makedirs(output_dir, exist_ok=True)
fname = os.path.join(output_dir, f"Y_{dataset}_{varname}_{'_'.join(str(m) for m in months)}_{op}.tsv")

with open(fname, "w") as f:
    f.write(header)
    for bloque in bloques:
        f.write(bloque)

print(f"Archivo guardado en: {fname}")
```
